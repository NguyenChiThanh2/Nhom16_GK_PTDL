```python
import pandas as pd
df = pd.read_csv('data_RFM.csv')
df
```

```python
# Đặt tên 11 phân khúc khách hàng
segment_names = ["Champions", "Loyal Customers", "Potential Loyalist",
                 "New Customers", "Promising", "Need Attention", 
                 "About To Sleep", "At Risk", "Can't Lose Them", "Lost"]

# Định nghĩa hàm phân khúc
def rfm_segment(row):
    recency_lower = [4, 2, 3, 4, 3, 2, 2, 1, 1, 1]
    recency_upper = [5, 5, 5, 5, 4, 3, 3, 2, 1, 2]
    frequency_lower = [4, 3, 1, 1, 1, 2, 1, 2, 4, 1]
    frequency_upper = [5, 5, 3, 1, 1, 3, 2, 5, 5, 2]
    monetary_lower = [4, 3, 1, 1, 1, 2, 1, 2, 4, 1]
    monetary_upper = [5, 5, 3, 1, 1, 3, 2, 5, 5, 2]

    for i in range(len(segment_names)):
        condition = (
            (row['RFM_Recency'] >= recency_lower[i]) & (row['RFM_Recency'] <= recency_upper[i]) &
            (row['RFM_Frequency'] >= frequency_lower[i]) & (row['RFM_Frequency'] <= frequency_upper[i]) &
            (row['RFM_Monetary'] >= monetary_lower[i]) & (row['RFM_Monetary'] <= monetary_upper[i])
        )
        if condition:
            return segment_names[i]
    return 'Other'

# Áp dụng hàm rfm_segment để gán phân khúc
df['Segment'] = df.apply(rfm_segment, axis=1)

# Hiển thị kết quả
print(df)
```

```python
df.to_csv('data_RFM_segments.csv', index=False)
```
